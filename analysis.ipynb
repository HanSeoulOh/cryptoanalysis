```python
import glob
import os
```

```python
os.chdir("./[Obfuscated] ExcelBinOutput")
jsons = []
for file in glob.glob("*.json"):
    jsons.append(file)
```

```python
import json
keys = set()
for file in jsons:
    with open(file,encoding='utf-8',) as jsonFile:
        data = json.load(jsonFile)
        for x in data:
            for y in x.keys():
                keys.add(y)
```

```python
import seaborn as sns
# keys
```

```python
keylist = list(keys)
```

```python
import string
count = {}
for i in range(11):
    count[str(i)] = {}
    for char in string.ascii_uppercase:
        count[str(i)][char] = 0
for key in keylist:
    for char, i in zip(key, range(11)):
        try:
            count[str(i)][char] += 1
        except:
            
            print("error on", key)
            break
```

```python
import pandas as pd
df = pd.DataFrame(count)
```

```python
df = df.reset_index().melt(id_vars=["index"], value_name = "occurrences", var_name="location")
```

```python
df = df.rename(columns={"index": "letter"})
```

```python
df2 = df[df.location == "0"][["letter", "occurrences"]]
sns.catplot(data=df, x="letter", y="occurrences", col="location", kind="bar", col_wrap=3)
```
